```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Data Indeks Standar Pencemar Udara (ISPU) di Provinsi DKI Jakarta ).xls.xlsx')
df.head(5)
```

```python
# Definisikan nilai missing yang kemungkinan terjadi
missing_values = ['', ' ', 'NaN', 'Nan', 'nan', '.', ',','-','--','---', 'TIDAK ADA DATA', 'KOSONG']
col_names = list(df.columns)
df[col_names] = df[col_names].replace(missing_values, np.nan)

# Hapus data kosong
df = df.dropna()
```

```python
df.head(5)
df.to_csv('clean_data.csv')
```
